```python
import cv2
import matplotlib.pyplot as plt

img_bgr = cv2.imread("../data/benchmarks/neonlady.png")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)


plt.subplot(1, 2, 1) # plt.subplot(rows, columns, index)
plt.imshow(img_bgr)
plt.title("BGR Image")
plt.axis('off')

plt.subplot(1, 2, 2) # 1 total rows, 2 total columns, 2 current position
plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis('off')

plt.show()
```

```python
img_grayscale = cv2.imread("../data/benchmarks/neonlady.png", cv2.IMREAD_GRAYSCALE)
# or we can use cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_grayscale, cmap="gray")
plt.title("Grayscale image")
plt.axis('off')

plt.show()

print(f"Original rgb image's shape: {img_rgb.shape}")
print(f"Grayscale image's shape:  {img_grayscale.shape}")
```

```python
_, img_binary = cv2.threshold(img_grayscale, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(img_grayscale, cmap='gray')
plt.title("Grayscale (0-255)")
plt.axis('off')

plt.subplot(122)
plt.imshow(img_binary, cmap='gray')
plt.title("Binary (0 or 255)")
plt.axis('off')

plt.show()
```

```python
_, img_binary_1 = cv2.threshold(img_grayscale, 100, 255, cv2.THRESH_BINARY)
_, img_binary_2 = cv2.threshold(img_grayscale, 150, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(img_binary_1, cmap='gray')
plt.title("Binary image - 100")
plt.axis('off')

plt.subplot(122)
plt.imshow(img_binary_2, cmap='gray')
plt.title("Binary image - 150")
plt.axis('off')

plt.show()
```
