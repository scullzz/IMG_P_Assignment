```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../data/benchmarks/coins.png")

kernel_3x3 = np.ones((3, 3), np.float32) / 9
blurred_image_3x3 = cv2.filter2D(image, -1, kernel_3x3)

kernel_5x5 = np.ones((5, 5), np.float32) / 25
blurred_image_5x5 = cv2.filter2D(image, -1, kernel_5x5)

kernel_9x9 = np.ones((9, 9), np.float32) / 81
blurred_image_9x9 = cv2.filter2D(image, -1, kernel_9x9)

plt.figure(figsize=(15, 7))

plt.subplot(1,4,1)
plt.imshow(image[...,::-1])
plt.title("Original Image")

plt.subplot(1,4,2)
plt.imshow(blurred_image_3x3[...,::-1])
plt.title("Blurred image with 3 by 3 kernel")

plt.subplot(1,4,3)
plt.imshow(blurred_image_5x5[...,::-1])
plt.title("Blurred image with 5 by 5 kernel")

plt.subplot(1,4,4)
plt.imshow(blurred_image_9x9[...,::-1])
plt.title("Blurred image with 9 by 9 kernel")

plt.show();
```

```python
image = cv2.imread("../data/benchmarks/neonlady.png")

blurred_image_sigma1 = cv2.GaussianBlur(image, (5, 5), sigmaX=0)
blurred_image_sigma5 = cv2.GaussianBlur(image, (5, 5), sigmaX=5)
blurred_image_sigma15 = cv2.GaussianBlur(image, (5, 5), sigmaX=15)

plt.figure(figsize=(15, 8))

plt.subplot(1,4,1)
plt.imshow(image[...,::-1])
plt.title("Original image")

plt.subplot(1,4,2)
plt.imshow(blurred_image_sigma1[...,::-1])
plt.title("Gaussian Blur, Sigma 1")

plt.subplot(1,4,3)
plt.imshow(blurred_image_sigma5[...,::-1])
plt.title("Gaussian Blur, sigma 5")

plt.subplot(1,4,4)
plt.imshow(blurred_image_sigma15[...,::-1])
plt.title("Gaussian Blur, sigma 15")

plt.show();
```

# Difference between Gaussian Blur and Uniform Blur

Uniform blur uses kernel who's weights are the same across the whole kernel, which creates square like structure on the edges and noisy part, while Gaussian blur uses averaged weights the following way: The higeshest priority has a center and it's neighbors has less values the farther away they are, which causes more smoother blur across the image.

```python
image = cv2.imread("../data/benchmarks/grass.png")

kernel_3x3 = np.ones((3, 3), np.float32) / 9

kernel_edge = np.array(
                [[-1, -1, -1],
                [-1,  8, -1],
                [-1, -1, -1]], dtype=np.float32)

kernel_sharpening = np.array(
                     [[ 0,-1, 0],
                      [-1, 5,-1],
                      [ 0,-1, 0]], dtype=np.float32)


output_image1 = cv2.filter2D(image, -1, kernel_3x3)
output_image2 = cv2.filter2D(image, -1, kernel_edge)
output_image3 = cv2.filter2D(image, -1, kernel_sharpening)


plt.figure(figsize=(15, 8))

plt.subplot(1,4,1)
plt.imshow(image[...,::-1])
plt.title("Original image")

plt.subplot(1,4,2)
plt.imshow(output_image1[...,::-1])
plt.title("Normal 3 by 3 kernel")

plt.subplot(1,4,3)
plt.imshow(output_image2[...,::-1])
plt.title("Kernel edge")

plt.subplot(1,4,4)
plt.imshow(output_image3[...,::-1])
plt.title("Sharpening kernel")

plt.show();
```
