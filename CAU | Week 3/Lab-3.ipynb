```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../data/benchmarks/lady.png")
print(image.shape)

transalation_matrix = np.float32([[1, 0, 100], [0, 1, 50]])

transalated_image = cv2.warpAffine(image, transalation_matrix, (image.shape[1], image.shape[0]))
print(transalated_image.shape)

plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
plt.imshow(image[...,::-1])
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(transalated_image[...,::-1])
plt.title("Transitioned image")

plt.show();
```

```python
transalation_matrix = np.float32([[1, 0, -100], [0, 1, -50]])
transalated_image = cv2.warpAffine(image, transalation_matrix, (image.shape[1], image.shape[0]))

plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
plt.imshow(image[...,::-1])
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(transalated_image[...,::-1])
plt.title("Transitioned image")

plt.show();
```

```python
(h, w) = image.shape[:2]
center = (h // 2, w // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))

rotation_matrix1 = cv2.getRotationMatrix2D(center, 90, 2.0)
rotated_image1 = cv2.warpAffine(image, rotation_matrix1, (w, h))


plt.figure(figsize=(15, 5))

plt.subplot(1,3,1)
plt.imshow(image[...,::-1])
plt.title("Original image")

plt.subplot(1,3,2)
plt.imshow(rotated_image[...,::-1])
plt.title("Rotated image 45")

plt.subplot(1,3,3)
plt.imshow(rotated_image1[...,::-1])
plt.title("Rotated image 90")

plt.show();
```

```python
image = cv2.imread("../data/benchmarks/camera.png")
h, w = image.shape[:2]
Factor = 2

scalling_matrix = cv2.getRotationMatrix2D((0, 0), 0, Factor)

scaled_image = cv2.warpAffine(image, scalling_matrix, (w*Factor, h*Factor))

print(f"Original shape: {image.shape}")
print(f"Scaled shape:   {scaled_image.shape}")

plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
plt.imshow(image[...,::-1])
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(scaled_image[...,::-1])
plt.title("Scalled image by 2x")

plt.show();
```

```python
image = cv2.imread("../data/benchmarks/coffee.png")
h, w = image.shape[:2]

Factors = [0.5, 1.5, 5]

plt.figure(figsize=(15, 3))

for i, factor in enumerate(Factors):
    scalling_matrix = cv2.getRotationMatrix2D((0, 0), 0, factor)
    scaled_image = cv2.warpAffine(image, scalling_matrix, (int(w*factor),int(h*factor)))

    plt.subplot(1, 3, i+1)
    plt.imshow(scaled_image[...,::-1])
    plt.title(f"Image scalled to the factor {factor}")

plt.show();
    
    
```

```python
shear_factor = 0.5

shear_matrix_x = np.float32([[1, shear_factor, 0], [0, 1, 0]])
sheared_image_x = cv2.warpAffine(image, shear_matrix_x, (w, h))

shear_matrix_y = np.float32([[1, 0, 0], [shear_factor, 1, 0]])
sheared_image_y = cv2.warpAffine(image, shear_matrix_y, (w, h))

plt.figure(figsize=(15, 5))

plt.subplot(1,3,1)
plt.imshow(image[...,::-1])
plt.title("Original image")

plt.subplot(1,3,2)
plt.imshow(sheared_image_x[...,::-1])
plt.title("Sheared image by factor 0.5 x")

plt.subplot(1,3,3)
plt.imshow(sheared_image_y[...,::-1])
plt.title("Sheared image by factor 0.5 y")

plt.show();
```
