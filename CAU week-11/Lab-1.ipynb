```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = np.zeros((12, 12), dtype=np.uint8)
img[3:9, 3:9] = 1
img[5, 5] = 0
img[1, 10] = 1

print(img)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(img, cmap='gray', vmin=0, vmax=1)
plt.title("Original Binary Image")
plt.axis("off")
plt.show()
```
