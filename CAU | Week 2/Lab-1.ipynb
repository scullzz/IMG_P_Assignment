```python
import cv2
import matplotlib.pyplot as plt

img_bgr = cv2.imread('../data/benchmarks/cat.png')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

subtracted_img = cv2.subtract(img_rgb, 100)

plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(122)
plt.imshow(subtracted_img)
plt.title("Subtracted Image")
plt.axis('off')

plt.show()
```

```python
values = [40, 80, 150]
plt.figure(figsize=(15, 5))

for i, value in enumerate(values):
    brightened_img = cv2.add(img_rgb, value)

    plt.subplot(1, 3, i+1)
    plt.imshow(brightened_img)
    plt.title(f"Added value {value}")
    plt.axis('off')

plt.show()
```

```python
multiplied_img = cv2.multiply(img_rgb, 2)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(multiplied_img)
plt.title("Image multiplied by 2")
plt.axis('off')

plt.show()
```

```python
values = [0.2, 0.5]

plt.figure(figsize=(10, 5))

for i, value in enumerate(values):
    mtp_img = cv2.multiply(img_rgb, value)

    plt.subplot(1, 2, i+1)
    plt.imshow(mtp_img)
    plt.title(f"Multiplied image by a factor of {value}")
    plt.axis('off')

plt.show()
```

```python
values = [2, 3.0, 5.0]

plt.figure(figsize=(10, 5))

for i, value in enumerate(values):
    d_img = cv2.divide(img_rgb, value)

    plt.subplot(1, 3, i+1)
    plt.imshow(d_img)
    plt.title(f"Divided image by a factor {value}")
    plt.axis('off')

plt.show()
```
